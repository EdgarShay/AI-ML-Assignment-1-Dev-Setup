Data Visualization: Create at least three informative visualizations:
A histogram for one numerical feature (e.g., petal length).

A box plot to show the distribution and potential outliers for one feature, grouped by the flower class (species).

A scatter plot of two features (e.g., petal length vs. petal width), with points colored by the flower species, to visually assess separability.

```python
# Import the seaborn library
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows
iris.head()
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Create a figure with 3 subplots arranged vertically
fig, axes = plt.subplots(3, 1, figsize=(12, 18))
```

```python
# 1. Histogram for petal length
sns.histplot(data=iris, x='petal_length', kde=True, bins=20, color='skyblue', ax=axes[0])
axes[0].set_title('Histogram of Petal Length', fontsize=16)
axes[0].set_xlabel('Petal Length (cm)', fontsize=14)
axes[0].set_ylabel('Frequency', fontsize=14)
```

```python
# 2. Box plot of sepal width grouped by species
sns.boxplot(data=iris, x='species', y='sepal_width', ax=axes[1])
axes[1].set_title('Box Plot of Sepal Width by Species')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Sepal Width (cm)')
```

```python
# 3. Scatter plot of petal length vs petal width, colored by species
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=axes[2])
axes[2].set_title('Scatter Plot: Petal Length vs Petal Width')
axes[2].set_xlabel('Petal Length (cm)')
axes[2].set_ylabel('Petal Width (cm)')
```

```python
# Adjust layout
plt.tight_layout()
plt.show()
```
